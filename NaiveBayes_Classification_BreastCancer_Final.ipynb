```python
import pandas as pd
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from sklearn import metrics as sm
from sklearn.naive_bayes import GaussianNB 
# GaussianNB as Independent Variables: Continuous Numeric #and target = Categorical
```

```python
bc_df=pd.read_excel('breast_cancer.xlsx',header=None)
bc_df
```

```python
X=bc_df.iloc[:,0:30] 
y=bc_df.iloc[:,30]
```

```python
X.head()
```

```python
y.head()
```

```python
X_s=scale(X)
X_s
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_s,y,test_size=0.2,random_state=15)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

GaussianNB implements the Gaussian Naive Bayes algorithm for classification. The likelihood of the features is assumed to be Gaussian:

```python
gNB=GaussianNB().fit(X_train,y_train)
gNB
```

```python
pred=gNB.predict(X_test)
pred
```

```python
#Test Accuracy
sm.accuracy_score(y_test,pred)
```

```python
#Training Accuracy
pred=gNB.predict(X_train)
pred
```

```python
sm.accuracy_score(y_train,pred)
```
